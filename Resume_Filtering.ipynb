**Exploding skills column**

```python
import pandas as pd
df = pd.read_csv('Resume_dataset.csv')
df['skills'] = df['skills'].str.strip('[]').str.split(',')
new_df1 = df.explode('skills')
new_df1['skills']  = new_df1['skills'].apply(str) 
new_df1['skills'] = new_df1['skills'].str.replace("'", '')
```

```python
new_df1
```

**Describing skills column after exploding skill column**

```python
new_df1['skills'].describe()
```

**Value counts of skills column after exploding skill column**

```python
new_df1['skills'].value_counts()
```

**Count plot of skills column**

```python
plt.figure(figsize=(50,50))
plt.xticks(rotation=90)
ax=sb.countplot(x= new_df1['skills'].value_counts() , data=new_df1)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.01 , p.get_height() * 1.01))
plt.grid()
```
